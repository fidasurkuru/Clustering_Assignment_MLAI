Apply PCA and reduce the features to 25 to 30 and then repeat the clustering

```python
# Apply PCA
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
# Determine number of components to keep
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
num_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print('Number of components to keep:', num_components)
```

```python
# Reduce number of features
X_pca = X_pca[:, :30]
```

```python
# Build KMeans model
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X_pca)
```

```python

# Compute silhouette score for KMeans
kmeans_silhouette = silhouette_score(X_pca, kmeans_labels)
print('KMeans Silhouette Score:', kmeans_silhouette)
```

```python
# Build Agglomerative model
agg = AgglomerativeClustering(n_clusters=5)
agg_labels = agg.fit_predict(X_pca)
```

```python
# Compute silhouette score for Agglomerative
agg_silhouette = silhouette_score(X_pca, agg_labels)
print('Agglomerative Silhouette Score:', agg_silhouette)
```
